```python
plt.figure(figsize=(10, 3))

# Plotting the lines
plt.plot(df_seasons['day_of_week'], df_seasons['mape_summer'], label='Summer [07/24 - 07/30]', color="#ffd11a")
plt.plot(df_seasons['day_of_week'], df_seasons['mape_fall'], label='Fall [10/09 - 10/15]', color="#ffa31a")
plt.plot(df_seasons['day_of_week'], df_seasons['mape_winter'], label='Winter [12/11 - 12/17]', color="#995c00")

# Adding labels and legend
plt.xlabel('Day of Week')
plt.ylabel('MAPE')
plt.title('Comparing CISO Daily MAPEs Over 3 Seasons ')
plt.legend()

# Display the plot
plt.show()
```

```python
df_mapes = pd.read_csv('gdrive/My Drive/CarbonPred/Data for Paper/section2_csvs_and_graphs/ALL_compare_daily_mapes.csv')
df_mapes['datetime'] = pd.to_datetime(df_mapes['datetime'])
# Adjust these for time range you want
df_mapes = df_mapes[df_mapes['datetime'] >= '2022-11-01 00:00:00']
df_mapes = df_mapes[df_mapes['datetime'] <= '2022-11-30 00:00:00']
```

```python
df_mapes
```

```python
plt.figure(figsize=(10, 3))

# Plotting the lines
plt.plot(df_mapes['datetime'], df_mapes['mape_CISO'], label='CISO', color="#ffa31a")
plt.plot(df_mapes['datetime'], df_mapes['mape_ERCO'], label='ERCO', color="#1aa3ff")
plt.plot(df_mapes['datetime'], df_mapes['mape_ISNE'], label='ISNE', color="#8533ff")

# Adding labels and legend
plt.xlabel('Date')
plt.ylabel('MAPE')
plt.title('Daily MAPE Across US Regions Over November 2022')
plt.legend()

# Display the plot
plt.show()
```

```python
dfs = pd.read_csv('gdrive/My Drive/CarbonPred/Data for Paper/section2_csvs_and_graphs/ALL_compare_1_to_4_step_mape.csv')
```

```python
dfs
```

```python
regions = dfs['region']
avg_1_step_mape = [dfs['avg_1_step_mape'][0], dfs['avg_1_step_mape'][1], dfs['avg_1_step_mape'][2]]
avg_2_step_mape = [dfs['avg_2_step_mape'][0], dfs['avg_2_step_mape'][1], dfs['avg_2_step_mape'][2]]
avg_3_step_mape = [dfs['avg_3_step_mape'][0], dfs['avg_3_step_mape'][1], dfs['avg_3_step_mape'][2]]
avg_4_step_mape = [dfs['avg_4_step_mape'][0], dfs['avg_4_step_mape'][1], dfs['avg_4_step_mape'][2]]

# create the dataframe
df = pd.DataFrame({'1-step pred': avg_1_step_mape, '2-step pred': avg_2_step_mape,
                   '3-step pred': avg_3_step_mape, '4-step pred': avg_4_step_mape,
                   }, index=regions)

# plot
ax = df.plot(kind='bar', figsize=(10, 3), rot=0, title='Average Daily MAPE Per Region For 4 Autoregressive Steps', ylabel='MAPE')
plt.show()
```
